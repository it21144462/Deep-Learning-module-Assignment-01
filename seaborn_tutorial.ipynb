```python
# Create random arrays from exponential, uniform, and normal distributions
exp_data = np.random.exponential(scale=1.0, size=100000)
uni_data = np.random.uniform(size=100000)
norm_data = np.random.normal(size=100000)

print(exp_data)
print(uni_data)
print(norm_data)

# Plot the histograms
plt.hist(exp_data, density=True, bins=100, histtype="step", color="blue", label="Exponential")
plt.hist(uni_data, density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(norm_data, density=True, bins=100, histtype="step", color="red", label="Normal")
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create meshgrid for X and Y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z
Z = X**2 + Y**2

# Plot the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('3D plot of Z = X^2 + Y^2')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
# Example: data = sns.load_dataset('your_dataset_name')
# Assuming the dataset is loaded and the features are: HP, Attack, Defense, Sp. Atk, Sp. Def, Speed

# For demonstration, I'll create a random dataset with similar structure
data = pd.DataFrame({
    'HP': np.random.randint(50, 150, 100),
    'Attack': np.random.randint(50, 150, 100),
    'Defense': np.random.randint(50, 150, 100),
    'Sp. Atk': np.random.randint(50, 150, 100),
    'Sp. Def': np.random.randint(50, 150, 100),
    'Speed': np.random.randint(50, 150, 100)
})

# Calculate Pearson correlation
pearson_corr = data.corr(method='pearson')

# Calculate Spearman rank correlation
spearman_corr = data.corr(method='spearman')

# Plot heatmaps
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# Pearson correlation heatmap
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', ax=ax[0])
ax[0].set_title('Pearson Correlation')

# Spearman rank correlation heatmap
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', ax=ax[1])
ax[1].set_title('Spearman Rank Correlation')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
url = '/content/drive/MyDrive/DL/Au_nanoparticle_dataset.csv'
df = pd.read_csv(url)

# Step 1: Create a new dataframe by filtering the specified columns
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

# Step 2: Display the first 20 samples
print(new_df.head(20))

# Step 3: Calculate the mean, standard deviation, and quartile values for each feature
mean_values = new_df.mean()
std_values = new_df.std()
quartiles = new_df.quantile([0.25, 0.5, 0.75])

print("Mean values:\n", mean_values)
print("\nStandard deviation values:\n", std_values)
print("\nQuartile values:\n", quartiles)

# Step 4: Plot histograms
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

for i, column in enumerate(new_df.columns):
    axes[i].hist(new_df[column], bins=30, alpha=0.7)
    axes[i].set_title(f'Histogram of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Step 5: Visualize scatter plots and histograms using pairplot
sns.pairplot(new_df)
plt.show()

# Step 6: Custom PairGrid plot
g = sns.PairGrid(new_df)
g.map_diag(sns.histplot, kde=True)  # Histograms with KDE on the diagonal
g.map_lower(sns.kdeplot, cmap='Blues', fill=True)  # Bivariate KDE on the lower half
g.map_upper(sns.scatterplot)  # Scatter plots on the upper half

plt.show()
```
